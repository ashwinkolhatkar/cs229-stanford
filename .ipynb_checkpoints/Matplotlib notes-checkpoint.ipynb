```python
import matplotlib.pyplot as plt
%matplotlib inline

x = [1,2,3,4,5,6,7]
y = [32, 34,25, 34,44,23,25]


plt.xlabel('Day of the week')
plt.ylabel('Temperature')
plt.title("Weather Graph")
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')

```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = [1,2,3,4,5,6,7]
y = [32, 34,25, 34,44,23,25]


plt.plot(x,y,'g') #makes green markers for the data points. Many format strings are available

plt.plot(x,y, color='red', marker='D', linestyle='', markersize=20, alpha=0.5)
#you can also use the hexadecimal code for colors
#alpha controls the transparency



```

```python
import matplotlib.pyplot as plt
%matplotlib inline

days = [1,2,3,4,5,6,7]
max_t = [52, 44,45, 44,44,38,32]
min_t = [32, 24,25, 24,24,23,25]

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')

plt.legend(loc = 'upper right', fontsize='small')

#keep checking the documentation to find more cool features

plt.grid()
```
